```python
import numpy as np
import pandas as pd
data = pd.read_excel(r'/home/bek/Documents/ОАВР/Official_Foreign_Exchange_Rates_NBRK_on_13_04_2020.xls')
ts = pd.DataFrame(data, columns=['USD']) # time series
ts = ts.values
ts = ts.reshape(-1)
time_series = ts
```

```python
l = len(time_series) // 2
ts = time_series[l:]
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
scores = np.empty(14)
for k in range(2, 16):
    X = time_series[:l - l%k]
    X = X.reshape(l//k, k)
    KMean = KMeans(n_clusters=k)
    KMean.fit(X)
    y_true = KMean.labels_
    X1 = ts[:l - l%k]
    X1 = X1.reshape(l//k, k)
    y_pred = KMean.predict(X1)
    scores[k - 2] = accuracy_score(y_true, y_pred)
k = 2 + np.argmax(scores)
print(max(scores),k)
```

```python
tau = 10
l = len(time_series) - tau + 1
X = np.empty(shape=(tau, l))
for i in range(tau):
    for j in range(l):
        X[i][j] = time_series[i + j]
print(X)
```

```python

C = X@X.transpose() / l
print(C)
```

```python
L, V = np.linalg.eigh(C)
print(L, end='\n\n')
print(V)
```

```python
r = 8
L_ = L[2:]
V_ = V[2:]
print(L_, end='\n\n')
print(V_)
```

```python
V_a = np.empty(shape=(tau-1,r)) # a - asterisk
for i in range(tau-1):
    for j in range(r):
        V_a[i][j] = V[i][j]
print(V_a.shape)
print(V_a)
```

```python
Q = np.array(time_series[l:])
print(len(Q))
```

```python
P = np.linalg.lstsq(V_a, Q)
P = P[0]
print(P)
```
